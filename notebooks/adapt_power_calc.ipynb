```python
import pandas as pd
import sys
import os
sys.path.append('C:\\Users\\szb37\\My Drive\\Projects\\ADAPT\\ADAPT codebase\\')
import src.config as config
import src.folders as folders
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 300  # Set display DPI


df = pd.read_csv(os.path.join(folders.power, f'adapt_power_avgs.csv'))
```

```python
fig, ax = plt.subplots(figsize=[9.6, 4.8],)
sns.lineplot(
    data = df,
    x = 'sample_size', 
    y = 'cgr_eqv',
    hue = 'scenario',
)

plt.title('CGR criteria', fontdict=config.title_fontdict)
ax.set_xlabel('Sample size', fontdict=config.axislabel_fontdict)
ax.set_xlim([50, 500])
ax.set_ylabel('Power to show equivalence', fontdict=config.axislabel_fontdict)
ax.set_ylim([0, 1.0])

plt.show()
```
