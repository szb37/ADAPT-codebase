```python
# Create x values from 0 to 30
x = np.linspace(0, 30, 1000)

# Calculate the normal distribution PDF with mean=15, sd=1
mean = 15
sd = 0.5
y = scipy.stats.norm.pdf(x, mean, sd)

# Create the plot using seaborn
fig = plt.figure(figsize=(8, 6))
sns.lineplot(x=x, y=y)
plt.xlabel('Psilocybin dose guess [mg]', fontdict=config.axislabel_fontdict)
plt.ylabel('Probability density', fontdict=config.axislabel_fontdict)
plt.xlim(0, 30)
plt.ylim(-0.025, 0.8)
plt.grid(True, alpha=0.3)

plt.axvline(x=15, color='red', linestyle='--', linewidth=1.5)
plt.axvline(x=mean-2*sd, color='gray', linestyle='--', linewidth=1.5)
plt.axvline(x=mean+2*sd, color='gray', linestyle='--', linewidth=1.5)

commons.Plots.save_fig(
    fig = fig,
    save_PNG = config.savePNG,
    save_SVG = config.saveSVG,
    dir_out = 'C:\\Users\\szb37\\My Drive\\Projects\\POP\\codebase\\exports POP\\blinding_quality\\discretize explainer',
    fname_out = 'conf_to_dose_max_conf')
```

```python
# Create x values from 0 to 30
x = np.linspace(0, 30, 1000)
mean = 15

# Define the standard deviations for each panel
sds = [0.5, 1.5, 2.5, 3.5, 4.5]

# Create figure with 5 panels in a single row
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

# Create each panel
for idx, sd in enumerate(sds):
    ax = axes[idx]
    
    # Calculate the normal distribution PDF
    y = scipy.stats.norm.pdf(x, mean, sd)
    
    # Plot the distribution
    sns.lineplot(x=x, y=y, ax=ax)
    ax.set_xlabel('Psilocybin dose guess [mg]', fontdict=config.axislabel_fontdict)
    #ax.set_ylabel('Probability density', fontdict=config.axislabel_fontdict)
    ax.set_xlim(0, 30)
    ax.set_ylim(-0.025, 0.8)

    ax.grid(True, alpha=0.3)

plt.tight_layout()
commons.Plots.save_fig(
    fig = fig,
    save_PNG = config.savePNG,
    save_SVG = config.saveSVG,
    dir_out = 'C:\\Users\\szb37\\My Drive\\Projects\\POP\\codebase\\exports POP\\blinding_quality\\discretize explainer',
    fname_out = 'conf_to_dose_slider')
```

```python
# Create figure with 4 panels in a single column
fig, axes = plt.subplots(4, 1, figsize=(6, 8,), dpi=300)

# Define confidence intervals for each panel (from top to bottom)
confidence_intervals = [
    [-0.5, 1.5],
    [0.5, 2],
    [2, 5],
    [0.5, 3.5]
]

# Create each panel
for idx, (ax, ci) in enumerate(zip(axes, confidence_intervals)):
    # Calculate mean of the confidence interval
    mean = (ci[0] + ci[1]) / 2
    
    # Set x-axis limits
    ax.set_xlim(-7, 7)
    
    # Draw horizontal confidence interval
    ax.plot(ci, [0, 0], linewidth=3, color='blue', marker='|', markersize=10, markeredgewidth=2)
    ax.plot(mean, 0, 'o', color='blue', markersize=8, zorder=5)

    # Remove y-axis
    ax.set_yticks([])
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    # Set x-axis label
    ax.set_xlabel('Outcome measure', fontdict=config.axislabel_fontdict)
    
    if idx in [0, 1]:
        ax.axvline(x=0, color='gray', linestyle='--', linewidth=1.5)
    else:
        ax.axvline(x=-4, color='gray', linestyle='--', linewidth=1.5)
        ax.axvline(x=4, color='gray', linestyle='--', linewidth=1.5)

    # Add grid
    ax.grid(True, alpha=0.3, axis='x')
    ax.set_ylim(-0.25, 0.25)

plt.tight_layout()
plt.show()

commons.Plots.save_fig(
    fig = fig,
    save_PNG = config.savePNG,
    save_SVG = config.saveSVG,
    dir_out = 'C:\\Users\\szb37\\My Drive\\Projects\\POP\\codebase\\exports POP\\blinding_quality\\discretize explainer',
    fname_out = 'hypothsis_vs_eq_testing')
```
