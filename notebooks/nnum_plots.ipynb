```python
import pandas as pd
import sys
import os
sys.path.append('C:\\Users\\szb37\\My Drive\\Projects\\ADAPT\\ADAPT codebase\\')
import src.config as config
import src.folders as folders
import matplotlib.pyplot as plt
import seaborn as sns
#plt.rcParams["figure.figsize"] = [16/4, 9/4] # Use this for displaying within notebook
plt.rcParams['figure.dpi'] = 300 # Set display DPI

df = pd.read_csv(os.path.join(folders.power, 'nnum_trialsResults.csv'))
```

```python
### CGR criteria
plt.rcParams['figure.dpi'] = 300 # Set display DPI

fig, ax = plt.subplots()
sns.lineplot(
    data = df,
    x = 'sample_size', 
    linewidth = 0.85,
    y = 'cgr_sd',
    hue = 'scenario',
)

plt.title('CGR', fontdict=config.nb_title)
ax.set_xlabel('Total sample size', fontdict=config.nb_axislabel)
ax.set_xlim([60, 300])
ax.set_ylabel('Ratio of trials with sig. unmasking', fontdict=config.nb_axislabel)
ax.set_ylim([0, 1.05])
ax.tick_params(axis='both', labelsize=config.nb_ticklabel)
plt.legend(
    title='Scenario', 
    title_fontproperties=config.nb_legendtitle, 
    bbox_to_anchor=(1.05, 1), 
    loc='upper left',
    fontsize = config.nb_legend)

plt.tight_layout()
plt.show()
```
