```python
import pandas as pd
import sys
import os
sys.path.append('C:\\Users\\szb37\\My Drive\\Projects\\ADAPT\\ADAPT codebase\\')
import src.config as config
import src.folders as folders
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 300  # Set display DPI
plt.style.use('seaborn-v0_8-notebook')  # notebook-optimized

df_power = pd.read_csv(os.path.join(folders.power, 'adapt_power.csv'))
```

```python
### CGR criteria
fig, ax = plt.subplots()
sns.lineplot(
    data = df_power,
    x = 'sample_size', 
    y = 'cgr_eqv',
    hue = 'scenario',
)

plt.title('CGR (ROPE: 0.14)', fontweight='bold', fontsize=18)
ax.set_xlabel('Total sample size', fontweight='bold')
ax.set_ylabel('Ratio of trials with equivalence', fontweight='bold')
plt.legend(title='Scenario', title_fontproperties={'weight': 'bold'})
ax.set_xlim([60, 400])
ax.set_ylim([0, 1.05])
plt.show()
```

```python
### BBI criteria (for the T arm)
df_tmp = df_power.copy()
df_tmp.sample_size = df_tmp.sample_size/2

fig, ax = plt.subplots()
sns.lineplot(
    data = df_tmp,
    x = 'sample_size', 
    y = 'bbi_T_eqv',
    hue = 'scenario',
)

plt.title('BBI of Treatment arm (ROPE: 0.2)', fontweight='bold', fontsize=18)
ax.set_xlabel('Sample size PER ARM', fontweight='bold')
ax.set_ylabel('Ratio of trials with equivalence', fontweight='bold')
plt.legend(title='Scenario', title_fontproperties={'weight': 'bold'})
ax.set_xlim([30, 200])
ax.set_ylim([0, 1.05])
plt.show()
```

```python
### Guessed mg WITHOUT confidence
fig, ax = plt.subplots()
sns.lineplot(
    data = df_power,
    x = 'sample_size', 
    y = 'gmg_eqv',
    hue = 'scenario',
)

plt.title('Guessed mg (ROPE: 5mg)', fontweight='bold', fontsize=18)
ax.set_xlabel('Total sample size', fontweight='bold')
ax.set_ylabel('Ratio of trials with equivalence', fontweight='bold')
plt.legend(title='Scenario', title_fontproperties={'weight': 'bold'})
ax.set_xlim([60, 400])
ax.set_ylim([0, 1.05])
plt.show()
```

```python
### Guessed mg WITH confidence
fig, ax = plt.subplots()
sns.lineplot(
    data = df_power,
    x = 'sample_size', 
    y = 'gmgc_eqv',
    hue = 'scenario',
)

plt.title('Guessed mg with confidence (ROPE: 5mg)', fontweight='bold', fontsize=18)
ax.set_xlabel('Total sample size', fontweight='bold')
ax.set_ylabel('Ratio of trials with equivalence', fontweight='bold')
plt.legend(title='Scenario', title_fontproperties={'weight': 'bold'})
ax.set_xlim([60, 400])
ax.set_ylim([0, 1.05])
plt.show()
```
